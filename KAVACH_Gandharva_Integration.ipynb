# KAVACH Gandharva Voice Detector Integration

This notebook demonstrates the integration and functionality of the `GandharvaVoiceDetector` module from the KAVACH-Sovereign-Intelligence project. It analyzes a sample audio file to determine if it's real or synthetic.


```
import sys
import os

# Add the KAVACH-Sovereign-Intelligence directory to the Python path
sys.path.append('.') # Assuming the notebook is in the KAVACH-Sovereign-Intelligence directory

from modules.akasha import AkashaLedger
from modules.gandharva import GandharvaVoiceDetector

# Initialize AkashaLedger (it creates akasha_ledger.json if it doesn't exist)
# Assuming akasha_ledger.json is in the same directory as the notebook relative to its execution context
ledger = AkashaLedger('akasha_ledger.json')

# Initialize GandharvaVoiceDetector
gandharva = GandharvaVoiceDetector(ledger=ledger)

# Define the path to the sample audio file
audio_file_path = 'recorded_sample.wav'

# Check if the audio file exists
if os.path.exists(audio_file_path):
    # Analyze the sample audio file
    print(f'
Analyzing audio file: {audio_file_path}')
    result = gandharva.analyze(path=audio_file_path, audio_id='recorded_sample.wav')
    print('
Gandharva Analysis Result:')
    print(result)
else:
    print(f'Error: Audio file not found at {audio_file_path}')
```
